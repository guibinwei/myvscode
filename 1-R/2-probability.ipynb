```r
b=100
k <- seq(0, b, by = 2)
y = rep(0.1,b/2 + 1)

n=30
k1 <- seq(0, n, by = 2)
y1=dbinom(k1, n, 0.5)

n=70
k2<- seq(0, n, by = 2)
y2=dbinom(k2, n, 0.5)

n=100
k3 <- seq(0, n, by = 2)
y3=dbinom(k3, n, 0.5)

plot(k, y,type = "l")
lines(k1, y1,type = "l")
lines(k2, y2,type = "l")
lines(k3, y3,type = "l")
```

```r
b=60
k <- 0:b
y = rep(0.1,b + 1)

n=20
k1 <- 0:n
y1=dbinom(k1, n, 0.5)

n=40
k2<- 0:n
y2=dbinom(k2, n, 0.5)

n=60
k3 <- 0:n
y3=dbinom(k3, n, 0.5)

plot(k, y,type = "l")
lines(k1, y1,type = "l", lwd=2, col = "red")
lines(k2, y2,type = "l", lwd=2, col = "green")
lines(k3, y3,type = "l", lwd=2, col = "blue")
```

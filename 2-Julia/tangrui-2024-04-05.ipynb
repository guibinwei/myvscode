```julia
using Plots
```

```julia
L = 100 #初始矩阵大小，对应L*L矩阵
```

```julia
b = 1.2
alpha =  0.3 
theta = 20 
K = 1 
H = 10 #额外阈值
```

```julia
A_mtx=0:0.1:20
```

```julia
A_gap=theta  .- A_mtx
```

```julia
K_A= K .* A_gap ./ theta
```

```julia
dR=-2   #dR=Ri-Rj 此处假设 对方的收益更多，测试压力对于改变策略的影响
```

```julia
aR= dR ./  K_A
```

```julia
P2 =1 ./ ( 1 .+ exp.(aR) )
```

```julia
plot(A_mtx, P2)
```

```julia
plot(K_A, P2)
```
